```python
from parksim.agents.rule_based_stanley_vehicle import RuleBasedStanleyVehicle
from parksim.vehicle_types import VehicleBody, VehicleConfig
from parksim.intent_predict.cnn.visualizer.instance_centric_generator import InstanceCentricGenerator
from parksim.pytypes import Position
import yaml
from yaml.loader import SafeLoader
from pathlib import Path 
import os
import numpy as np
```

```python
MAP_DIMS = (140, 70)
def get_agent_in_rand_position(agent_id): 
    body = VehicleBody()
    config = VehicleConfig()
    vehicle = RuleBasedStanleyVehicle(agent_id, body, config)
    vehicle.state.x = Position(np.random.rand() * MAP_DIMS[0], np.random.rand() * MAP_DIMS[1], 0)
    return vehicle
```

```python
vehicle = get_agent_in_rand_position(0)
generator = InstanceCentricGenerator()
all_agents = {i : get_agent_in_rand_position(i) for i in range(1, 10)}

history = [[agent.get_state_dict() for agent in all_agents.values()]]
```

```python
generator.inst_centric(vehicle, all_agents, history)
```

```python
generator.plot_frame(all_agents)
```
